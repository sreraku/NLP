```python
import pandas as pd
import numpy as np
import re
import string
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# from google.colab import files
from translate import Translator
import time
```

```python
severe = pd.read_csv("https://adamhyman-public.s3.amazonaws.com/w266/severe.csv")
threat1 = pd.read_csv("https://adamhyman-public.s3.amazonaws.com/w266/threat.csv")

threat2 = threat1[353:370].copy()
threat3 = threat1.copy()

print (severe.shape)
print (threat1.shape)
print (threat2.shape)
print (threat3.shape)

display (threat2)
```

```python
english_chinese_translator = Translator(from_lang='en', to_lang='zh')
chinese_english_translator = Translator(from_lang='zh', to_lang='en')

english_french_translator = Translator(from_lang='en', to_lang='fr')
french_english_translator = Translator(from_lang='fr', to_lang='en')

english_hindi_translator = Translator(from_lang='en', to_lang='hi')
hindi_english_translator = Translator(from_lang='hi', to_lang='en')




def back_translate_chinese (english_comment):
  chinese_comment = english_chinese_translator.translate(english_comment[:499])
  time.sleep(1)
  english_comment_back = chinese_english_translator.translate(chinese_comment[:499])
  time.sleep(1)
  return english_comment_back

def back_translate_french (english_comment):
  french_comment = english_french_translator.translate(english_comment[:499])
  time.sleep(1)
  english_comment_back = french_english_translator.translate(french_comment[:499])
  time.sleep(1)
  return english_comment_back

def back_translate_hindi (english_comment):
  hindi_comment = english_hindi_translator.translate(english_comment[:499])
  time.sleep(1)
  english_comment_back = hindi_english_translator.translate(hindi_comment[:499])
  time.sleep(1)
  return english_comment_back
```

```python
for i in range(1):
  print ('Original text:')
  print (threat1['comment_text'][i])
  print ('\nBack translated from Chinese:')
  print (back_translate_chinese(threat1['comment_text'][i]))
  print ('\nBack translated from French:')
  print (back_translate_french(threat1['comment_text'][i]))
  print ('\nBack translated from Hindi:')
  print (back_translate_hindi(threat1['comment_text'][i]))
```

```python
threat2["comment_text"] = threat2["comment_text"][:499].apply(lambda text: back_translate_hindi(text))
```

```python
threat2.to_csv('threat2-hindi-dkjf4535s.csv', index=False)
```
