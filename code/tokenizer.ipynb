```python
from transformers import BertTokenizer
from transformers import XLNetTokenizer
from transformers import T5Tokenizer
```

```python
sample = 'i will be back in a few hours canuck to wreak havoc on your canadian fagot ass'
sample2 = 'sup you crackheads  just kidding  this is my user page here i say fart fuck shit damn ass nigger jew spick and kyke a lot  go fuck yourself if you do not like it'
```

```python
btokenizer = BertTokenizer.from_pretrained('bert-base-cased')
encoding = btokenizer.encode(sample)
print(encoding)
print(btokenizer.convert_ids_to_tokens(encoding))
print (btokenizer.convert_ids_to_tokens(btokenizer.encode(sample2)))
```

```python
t5_tokenizer = T5Tokenizer.from_pretrained('t5-base')
encoding = t5_tokenizer.encode(sample)
print(encoding)
print(t5_tokenizer.convert_ids_to_tokens(encoding))
print (t5_tokenizer.convert_ids_to_tokens(t5_tokenizer.encode(sample2)))
```

```python
Xtokenizer = XLNetTokenizer.from_pretrained("xlnet-base-cased")
encoding = Xtokenizer.encode(sample)
print(encoding)
print(Xtokenizer.convert_ids_to_tokens(encoding))
print (Xtokenizer.convert_ids_to_tokens(Xtokenizer.encode(sample2)))
```
